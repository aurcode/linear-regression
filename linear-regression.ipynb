```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston.DESCR)
```

Formula to minimize the mean square error (MSE): $\beta - (X^{T}X)^{-1}X^{T}Y$

```python
X = np.array(boston.data[:, 5])
# X is average number of rooms per dwelling
Y = np.array(boston.target)
# Y is house-price

plt.scatter(X, Y, alpha=0.15)

# We add a column of 1 for independent term.
X = np.array([np.ones(506), X]).T

B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.title('House price per number of rooms in Boston')
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('House-price')

plt.plot([4,9], [B[0] + B[1]*4, B[0] + B[1] * 9], c='red')
# plt.savefig('house-price-per-number-of-rooms.jpg', dpi=150)
plt.show()
```
